# Introducción a Pandas Series

Una `Serie` de Pandas es una estructura de datos unidimensional capaz de contener cualquier tipo de datos (enteros, cadenas, números de punto flotante, objetos de Python, etc.). Es muy similar a una columna en una tabla de hoja de cálculo o una base de datos, o un array de NumPy de una dimensión. La diferencia principal es que una Serie tiene un *índice* asociado a los datos, lo que permite un acceso y manipulación más flexibles.

Comencemos importando la librería `pandas`.

```python
import pandas as pd
import numpy as np

print("Pandas y NumPy importados correctamente.")
```

## Creación de Series

Se pueden crear Series de varias maneras:

### 1. Desde una lista de Python

```python
data = [10, 20, 30, 40, 50]
s = pd.Series(data)
display(s)
print(type(s))
```

Por defecto, Pandas asigna un índice numérico que comienza en 0.

### 2. Desde un array de NumPy

```python
arr = np.array([100, 200, 300, 400])
s_np = pd.Series(arr)
display(s_np)
```

### 3. Desde un diccionario de Python

Las claves del diccionario se convertirán en el índice de la Serie y los valores en los datos.

```python
data_dict = {'a': 1, 'b': 2, 'c': 3, 'd': 4}
s_dict = pd.Series(data_dict)
display(s_dict)
```

### 4. Con un índice personalizado

Podemos especificar un índice explícitamente al crear la Serie.

```python
data = [10, 20, 30, 40]
index = ['uno', 'dos', 'tres', 'cuatro']
s_custom_index = pd.Series(data, index=index)
display(s_custom_index)
```

## Acceso a elementos de una Serie

Los elementos de una Serie se pueden acceder de varias maneras, similar a los arrays de NumPy o los diccionarios de Python.

### 1. Por etiqueta de índice

```python
display(s_custom_index['uno'])

# Acceso a múltiples elementos por una lista de etiquetas
display(s_custom_index[['uno', 'tres']])
```

### 2. Por posición entera (iloc)

Podemos usar `iloc` para acceder a elementos por su posición numérica, como en una lista o array.

```python
display(s_custom_index.iloc[0]) # Primer elemento
display(s_custom_index.iloc[2]) # Tercer elemento

# Acceso a múltiples elementos por una lista de posiciones
display(s_custom_index.iloc[[0, 3]])
```

### 3. Slicing

El slicing funciona de forma similar a las listas y arrays, tanto por etiquetas como por posiciones.

```python
# Slicing por posición
display(s_custom_index[0:2]) # Elementos en posición 0 y 1

# Slicing por etiqueta (¡el final es inclusivo!)
display(s_custom_index['uno':'tres']) # Elementos desde 'uno' hasta 'tres' inclusive
```

## Atributos comunes de una Serie

Las Series tienen varios atributos útiles para inspeccionar sus características.

```python
s_example = pd.Series([10, 20, 30, 'cuarenta', 50.5], name='MiSerieEjemplo', index=['a', 'b', 'c', 'd', 'e'])
display(s_example)

print(f"Índice: {s_example.index.tolist()}")
print(f"Valores: {s_example.values}")
print(f"Tipo de dato (dtype): {s_example.dtype}")
print(f"Nombre de la Serie: {s_example.name}")
print(f"Número de elementos (size): {s_example.size}")
print(f"Forma (shape): {s_example.shape}") # Siempre es (n,) para una Serie
```

## Operaciones Básicas con Series

Las Series de Pandas permiten realizar operaciones matemáticas de forma vectorizada, similar a NumPy, y manejan automáticamente los índices.

### 1. Operaciones Aritméticas

```python
s1 = pd.Series([1, 2, 3, 4], index=['a', 'b', 'c', 'd'])
s2 = pd.Series([10, 20, 30, 40], index=['a', 'b', 'c', 'd'])

print("s1:"); display(s1)
print("s2:"); display(s2)

print("\nSuma (s1 + s2):"); display(s1 + s2)
print("Resta (s1 - s2):"); display(s1 - s2)
print("Multiplicación (s1 * s2):"); display(s1 * s2)
print("División (s1 / s2):"); display(s1 / s2)

# Las operaciones se alinean por índice. Si los índices no coinciden, se producirá NaN.
s3 = pd.Series([5, 6, 7], index=['b', 'c', 'e'])
print("\nSuma (s1 + s3) con índices no coincidentes:"); display(s1 + s3)
```

### 2. Indexación Booleana

Permite seleccionar elementos basándose en una condición lógica.

```python
s_numeric = pd.Series([10, 15, 20, 25, 30], index=['a', 'b', 'c', 'd', 'e'])
display(s_numeric)

# Seleccionar elementos mayores que 20
mask = s_numeric > 20
print("\nMáscara booleana:"); display(mask)
print("Elementos mayores que 20:"); display(s_numeric[mask])

# O de forma abreviada
print("\nElementos mayores que 20 (abreviado):"); display(s_numeric[s_numeric > 20])
```

### 3. Manejo de Valores Faltantes (NaN)

Las Series pueden contener valores `NaN` (Not a Number) para representar datos faltantes. Pandas proporciona métodos para identificarlos y manejarlos.

```python
s_nan = pd.Series([1, 2, np.nan, 4, np.nan, 6])
display(s_nan)

print("\n¿Valores nulos? (isnull):"); display(s_nan.isnull())
print("¿Valores no nulos? (notnull):"); display(s_nan.notnull())

print("\nEliminar valores nulos (dropna):"); display(s_nan.dropna())

print("\nRellenar valores nulos con 0 (fillna):"); display(s_nan.fillna(0))
print("\nRellenar valores nulos con el valor anterior (fillna method='ffill'):"); display(s_nan.fillna(method='ffill'))
print("\nRellenar valores nulos con el valor siguiente (fillna method='bfill'):"); display(s_nan.fillna(method='bfill'))
```

### 4. Estadísticas Descriptivas

Se pueden obtener rápidamente estadísticas resumidas de una Serie numérica.

```python
s_stats = pd.Series([10, 20, 30, 40, 50, 10, 20, 30, 40, 50, 60, 70])
display(s_stats)

print(f"Media: {s_stats.mean()}")
print(f"Mediana: {s_stats.median()}")
print(f"Desviación estándar: {s_stats.std()}")
print(f"Mínimo: {s_stats.min()}")
print(f"Máximo: {s_stats.max()}")

print("\nEstadísticas descriptivas completas (describe()):")
display(s_stats.describe())
```
