# Introducción a Matplotlib

Matplotlib es una biblioteca de visualización de datos en Python. Es muy versátil y permite crear una amplia gama de gráficos estáticos, animados e interactivos. Es la base de muchas otras bibliotecas de visualización en Python.

En este notebook, exploraremos cómo usar Matplotlib para crear diferentes tipos de gráficos con datos de ejemplo.

```python
# Importar las bibliotecas necesarias
import matplotlib.pyplot as plt
import numpy as np

print("Matplotlib y NumPy importados correctamente.")
```

## Generación de Datos de Ejemplo

Para nuestros ejemplos, vamos a generar algunos datos numéricos simples utilizando NumPy.

```python
# Datos para un gráfico de línea (series temporales o función)
x_line = np.linspace(0, 10, 100) # 100 puntos entre 0 y 10
y_line = np.sin(x_line) # Seno de x

# Datos para un gráfico de dispersión (relación entre dos variables)
x_scatter = np.random.rand(50) * 10
y_scatter = 2 * x_scatter + np.random.randn(50) * 5

# Datos para un gráfico de barras (valores categóricos)
categorias = ['A', 'B', 'C', 'D', 'E']
valores_barras = np.random.randint(1, 10, size=len(categorias))

# Datos para un histograma (distribución de una sola variable)
datos_hist = np.random.randn(1000) # 1000 números aleatorios de una distribución normal

print("Datos de ejemplo generados:")
print(f"x_line (primeros 5): {x_line[:5]}")
print(f"y_line (primeros 5): {y_line[:5]}")
print(f"x_scatter (primeros 5): {x_scatter[:5]}")
print(f"y_scatter (primeros 5): {y_scatter[:5]}")
print(f"Categorías: {categorias}")
print(f"Valores de barras: {valores_barras}")
print(f"datos_hist (primeros 5): {datos_hist[:5]}")
```

## 1. Gráfico de Línea Básico

Un gráfico de línea es ideal para mostrar la evolución de una variable a lo largo del tiempo o a través de un continuo. Es útil para identificar tendencias.

```python
# Crear un gráfico de línea
plt.figure(figsize=(10, 6)) # Define el tamaño de la figura
plt.plot(x_line, y_line) # Crea el gráfico de línea
plt.title('Gráfico de Línea Simple') # Añade un título
plt.xlabel('Eje X') # Etiqueta para el eje X
plt.ylabel('Eje Y') # Etiqueta para el eje Y
plt.grid(True) # Muestra una cuadrícula
plt.show() # Muestra el gráfico
```

## 2. Gráfico de Dispersión (Scatter Plot)

Los gráficos de dispersión son excelentes para visualizar la relación entre dos variables numéricas, ayudando a identificar correlaciones o patrones.

```python
# Crear un gráfico de dispersión
plt.figure(figsize=(10, 6))
plt.scatter(x_scatter, y_scatter, alpha=0.7) # Crea el gráfico de dispersión con transparencia
plt.title('Gráfico de Dispersión')
plt.xlabel('Variable X')
plt.ylabel('Variable Y')
plt.grid(True)
plt.show()
```

## 3. Gráfico de Barras (Bar Chart)

Los gráficos de barras son útiles para comparar cantidades entre diferentes categorías discretas.

```python
# Crear un gráfico de barras
plt.figure(figsize=(10, 6))
plt.bar(categorias, valores_barras, color='skyblue') # Crea el gráfico de barras
plt.title('Gráfico de Barras de Valores por Categoría')
plt.xlabel('Categoría')
plt.ylabel('Valor')
plt.grid(axis='y', linestyle='--') # Cuadrícula solo en el eje Y
plt.show()
```

## 4. Histograma

Los histogramas muestran la distribución de una única variable numérica, dividiendo los datos en 'bins' y contando cuántos valores caen en cada bin.

```python
# Crear un histograma
plt.figure(figsize=(10, 6))
plt.hist(datos_hist, bins=30, edgecolor='black', alpha=0.7) # Crea el histograma con 30 bins
plt.title('Histograma de Distribución de Datos')
plt.xlabel('Valor')
plt.ylabel('Frecuencia')
plt.grid(axis='y', linestyle='--')
plt.show()
```

## 5. Personalización de Estilos de Gráficos

Matplotlib ofrece muchas opciones para personalizar el aspecto de tus gráficos. Puedes cambiar colores, estilos de línea, marcadores, fuentes y más.

### 5.1 Gráfico de Línea con Estilo Personalizado

Vamos a modificar el gráfico de línea para cambiar el color, añadir marcadores y un estilo de línea diferente.

```python
# Crear un gráfico de línea con estilo personalizado
plt.figure(figsize=(12, 7))
plt.plot(x_line, y_line, color='red', linestyle='--', marker='o', markersize=4, label='Función Seno')
plt.title('Gráfico de Línea Personalizado')
plt.xlabel('Eje X')
plt.ylabel('Seno(X)')
plt.grid(True)
plt.legend() # Muestra la leyenda
plt.show()
```

### 5.2 Gráfico de Dispersión con Colores y Tamaño por Valor

Podemos usar una tercera variable para determinar el color o el tamaño de los puntos en un gráfico de dispersión.

```python
# Crear un gráfico de dispersión con colores y tamaño por valor
sizes = y_scatter**2 # Usaremos el cuadrado de y_scatter para el tamaño
colors = x_scatter # Usaremos x_scatter para mapear los colores

plt.figure(figsize=(12, 7))
scatter = plt.scatter(x_scatter, y_scatter, c=colors, s=sizes, cmap='viridis', alpha=0.8)
plt.title('Gráfico de Dispersión con Colores y Tamaños Variables')
plt.xlabel('Variable X')
plt.ylabel('Variable Y')
plt.colorbar(scatter, label='Valor de X') # Añadir barra de color
plt.grid(True)
plt.show()
```

### 5.3 Estilos Predefinidos de Matplotlib

Matplotlib incluye una serie de estilos predefinidos que puedes aplicar a tus gráficos para cambiar su apariencia global fácilmente.

```python
# Listar los estilos disponibles
print("Estilos disponibles:", plt.style.available)

# Aplicar un estilo diferente
plt.style.use('seaborn-v0_8-darkgrid') # Prueba con 'ggplot', 'fivethirtyeight', 'dark_background', etc.

plt.figure(figsize=(10, 6))
plt.plot(x_line, y_line, label='Datos Originales')
plt.title('Gráfico con Estilo "seaborn-v0_8-darkgrid"')
plt.xlabel('Eje X')
plt.ylabel('Seno(X)')
plt.legend()
plt.show()

# Volver al estilo por defecto para los siguientes gráficos
plt.style.use('default')
```

### 5.4 Gráfico de Barras con Anotaciones

Podemos añadir texto y flechas para resaltar puntos específicos en los gráficos.

```python
# Crear un gráfico de barras con anotaciones
plt.figure(figsize=(10, 6))
bars = plt.bar(categorias, valores_barras, color=['skyblue', 'lightcoral', 'lightgreen', 'gold', 'lightpink'])
plt.title('Gráfico de Barras con Anotaciones')
plt.xlabel('Categoría')
plt.ylabel('Valor')
plt.grid(axis='y', linestyle='--')

# Añadir etiquetas de valor en la parte superior de cada barra
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 0.5, round(yval, 2), ha='center', va='bottom')

# Anotar la barra con el valor más alto
max_value_index = np.argmax(valores_barras)
max_category = categorias[max_value_index]
max_value = valores_barras[max_value_index]

plt.annotate(f'Máximo: {max_value}', xy=(max_value_index, max_value), xytext=(max_value_index + 0.5, max_value + 2),
             arrowprops=dict(facecolor='black', shrink=0.05), fontsize=10, color='darkblue')

plt.show()
```

## 6. Múltiples Gráficos en una Misma Figura (Subplots)

Matplotlib permite organizar múltiples gráficos en una sola figura utilizando subplots. Esto es útil para comparar diferentes aspectos de los datos o para mostrar diferentes tipos de gráficos relacionados.

### 6.1 Crear Subplots con `plt.subplots()`

La función `plt.subplots()` es la forma más común y flexible de crear subplots. Devuelve una figura y un array de ejes.

```python
# Crear una figura con 1 fila y 2 columnas de subplots
fig, axes = plt.subplots(1, 2, figsize=(15, 6))

# Primer subplot: Gráfico de Línea
axes[0].plot(x_line, y_line, color='blue', linestyle='-', label='Seno(X)')
axes[0].set_title('Gráfico de Línea')
axes[0].set_xlabel('Eje X')
axes[0].set_ylabel('Eje Y')
axes[0].grid(True)
axes[0].legend()

# Segundo subplot: Gráfico de Dispersión
axes[1].scatter(x_scatter, y_scatter, color='green', alpha=0.7)
axes[1].set_title('Gráfico de Dispersión')
axes[1].set_xlabel('Variable X')
axes[1].set_ylabel('Variable Y')
axes[1].grid(True)

# Ajustar el diseño para evitar superposiciones
plt.tight_layout()
plt.show()
```

### 6.2 Subplots con Diferentes Tamaños (GridSpec)

Para diseños de subplots más complejos, donde los subplots tienen diferentes tamaños o abarcan varias filas/columnas, puedes usar `GridSpec`.

```python
from matplotlib.gridspec import GridSpec

fig = plt.figure(figsize=(16, 8))
gs = GridSpec(2, 2) # Una cuadrícula de 2x2

# Subplot superior izquierdo (ocupa 1 fila, 1 columna)
ax1 = fig.add_subplot(gs[0, 0])
ax1.hist(datos_hist, bins=20, color='purple', edgecolor='black', alpha=0.7)
ax1.set_title('Histograma de Datos')

# Subplot superior derecho (ocupa 1 fila, 1 columna)
ax2 = fig.add_subplot(gs[0, 1])
ax2.bar(categorias, valores_barras, color='orange')
ax2.set_title('Gráfico de Barras')

# Subplot inferior (ocupa 1 fila, 2 columnas)
ax3 = fig.add_subplot(gs[1, :]) # gs[1, :] significa todas las columnas de la fila 1
ax3.plot(x_line, y_line * 2, color='darkred', label='2 * Seno(X)') # Otro gráfico de línea
ax3.set_title('Gráfico de Línea Ampliado')
ax3.set_xlabel('Eje X')
ax3.set_ylabel('Eje Y')
ax3.grid(True)
ax3.legend()

plt.tight_layout()
plt.show()
```
